```python
import tkinter as tk
from tkinter import *
import random

#window
win=tk.Tk()
win.title("參數產生器")
win.geometry("600x800")
win.configure(bg='gray20')


#Label
titleLabel=Label(win,text="請輸入以下i j k 個數，以及想要產生的參數範圍",font=('微軟正黑體 bold',20),fg='steelblue2',bg='gray20')
titleLabel.place(x=300,y=80,anchor=N)

i_Label = Label(win,text="i",font=('微軟正黑體 bold',20),fg='steelblue2',bg='gray20')
i_Label.place(x=300,y=200,anchor=N)

j_Label = Label(win,text="j",font=('微軟正黑體 bold',20),fg='steelblue2',bg='gray20')
j_Label.place(x=300,y=300,anchor=N)

k_Label=Label(win,text="k",font=('微軟正黑體 bold',20),fg='steelblue2',bg='gray20')
k_Label.place(x=300,y=400,anchor=N)

rand1=Label(win,text="Random_Lower",font=('微軟正黑體 bold',20),fg='steelblue2',bg='gray20')
rand1.place(x=300,y=500,anchor=N)

rand2=Label(win,text="Random_Higher",font=('微軟正黑體 bold',20),fg='steelblue2',bg='gray20')
rand2.place(x=300,y=600,anchor=N)


#Entry
i_Entry = tk.Entry(win,font=('微軟正黑體 bold',14))
i_Entry.place(x=100,y=250,width=400,height=40)

j_Entry = tk.Entry(win,font=('微軟正黑體 bold',14))
j_Entry.place(x=100,y=350,width=400,height=40)

k_Entry = tk.Entry(win,font=('微軟正黑體 bold',14))
k_Entry.place(x=100,y=450,width=400,height=40)

rand1_Entry = tk.Entry(win,font=('微軟正黑體 bold',14))
rand1_Entry.place(x=100,y=550,width=400,height=40)

rand2_Entry = tk.Entry(win,font=('微軟正黑體 bold',14))
rand2_Entry.place(x=100,y=650,width=400,height=40)


def create_dict():
    i_value = int(i_Entry.get())
    j_value = int(j_Entry.get())
    k_value = int(k_Entry.get())
    rand1_value = int(rand1_Entry.get())
    rand2_value = int(rand2_Entry.get())
    
    dict_keys = []
    for i in range(1, i_value + 1):
        for j in range(1, j_value + 1):
            for k in range(1, k_value + 1):
                dict_keys.append(tuple([i,j,k]))
    dict_keys = tuple(dict_keys)
    
    dict_values = []
    for i in range(1, i_value + 1):
        for j in range(1, j_value + 1):
            for k in range(1, k_value + 1):
                dict_values.append(random.randint(rand1_value,rand2_value))
    print(dict(zip(dict_keys, dict_values)))
    return dict(zip(dict_keys, dict_values))           
        

def ggg():
    a = 100

bt_Send=tk.Button(win,text="送出",font=('微軟正黑體 bold',20),fg='gray10',bg='SeaGreen2',command=create_dict)
bt_Send.place(x=150,y=700,width=300,height=50)



win.mainloop()
```
